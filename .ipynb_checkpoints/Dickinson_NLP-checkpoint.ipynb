```python
import os
import re


# Script for preprocessing complete raw poems; I have to manually fix line breaks
# after this.

infile = "edCompleteRaw.txt"
outfile = "edCompleteFormatted.txt"
outfile2 = "edCompleteFormatted2.txt"

with open(infile, 'r') as raw:
    buff = raw.read().split()

with open(outfile2, 'w') as cleaned:
    
    # Split by poem number
    
    for word in buff:
        
        # Has poem number
        if re.search('[A-Za-z]*[0-9]+', word):
            
            first = ""
            poem_num = ""
            second = ""
            
            for letter in word:
                if re.match('[0-9]', letter):
                    poem_num += str(letter)
                elif len(poem_num) > 0:
                    second += letter
                else:
                    first += letter
            
            cleaned.write(first + "\n\n" + poem_num + "\n\n" + second + ' ')
        
        # Sometimes double quotation marks when they begin consecutive lines
        
        elif re.search('""', word):
            
            first = ""
            second = ""
            
            for letter in word:
                if letter == '"' and first[-1] == '"':
                    second += letter
                elif len(second) > 0:
                    second += letter
                else:
                    first += letter
                    
            cleaned.write(first + "\n" + second + ' ')
        
        # Could be combined word from line break
        # Check for capital letters
        
        elif re.search('[A-Za-z?!-,.;]+[A-Z]', word):
            
            first = ""
            second = ""
            
            for letter in word:
                if letter.isupper() and len(first) > 0:
                    second += letter
                elif len(second) > 0:
                    second += letter
                else:
                    first += letter
            
            cleaned.write(first + "\n" + second + ' ')
    
        # If word begins with capital letter, could be start
        # of new verse. Need to add newline, controlling only
        # for "I"
        
        elif word[0].isupper() and word != 'I':
            
            cleaned.write("\n" + word + ' ')
            
        else:
            cleaned.write(word + ' ')
```

```python
########################## GENERATING SEPARATE POEM FILES BY POEM NUMBER #################

import pickle
import json
import re
import string

inf = open("edCompleteNoDash.txt", 'r')

buff = ""
out = open("./edCompleteNoDash/0.txt", 'w') # dummy file to let loop work 

for line in inf:
    
    if re.match('[0-9]+', line[:-1]):  # line indicates new poem number
        
        out.write(buff) # store previous poem
        out.close()
        
        out = open("./edCompleteNoDash/{}.txt".format(int(line[:-1])), 'w')
        
        buff = line # buff now starts with new poem number
        
    else:
        
        buff = buff + line
        
out.write(buff)
out.close()
inf.close()

os.remove("./edCompleteNoDash/0.txt")  # kill dummy file
    
```

```python
################## GENERATING LOWERCASE POEM FILES WITH ONLY APOSTROPHES REMAINING ##################

import pickle
import json
import re
import string
import os


root = './edCompleteNoDash'
newdir = './edCompleteSimple'

buff = ""

for filename in os.listdir(root):
    
    inf = open(os.path.join(root, filename), 'r')
    out = open(os.path.join(newdir, filename), 'w')
    
    for line in inf:
        
        buff = line.replace("-", "")
        buff = line.replace("!", "")
        buff = buff.replace("?", "")
        buff = buff.replace(",", "")
        buff = buff.replace(".", "")
        buff = buff.replace(";", "")
        buff = buff.replace(":", "")
        buff = buff.replace("\"", "")
        buff = buff.lower()
        
        out.write(buff)
    
    inf.close()
    out.close()
    
```

```python
############ BASELINE POEM AFFECT SCORE CALCULATOR ###################

import pickle
import json
import os
from collections import defaultdict

with open("sent_dict.json", 'rb') as f:
    sent_dict = json.load(f)

root = './edCompleteSimple'

# k:v is poem number:(AS density per hitcount, AS density per wordcount)
# where hitcount = num words with scores found in sent_dict
# and wordcount = total words in poem

AS_dict = defaultdict(float)   
    
for filename in os.listdir(root):
    
    with open(os.path.join(root, filename), 'r') as inf:
        buff = inf.read().split()
        
    del buff[0]  # remove poem number
    
    hitcount = 0   # num words with score in sent_dict
    AS_total = 0
    
    for word in buff:
        
        if word in sent_dict.keys():
            
            hitcount += 1
            AS_total += sent_dict[word]
    
    AS_dict[filename[:-4]] = (AS_total/hitcount, AS_total/len(buff))


with open("AS_dict.json", 'w') as f:
    json.dump(AS_dict, f)
    
with open("AS_dict.pickle", 'wb') as f:
    pickle.dump(AS_dict, f)
```

```python
############### PRINT AS VALUES PER WORD IN A POEM #############

with open("./edCompleteSimple/1660.txt", 'r') as f:
    buff = f.read().split()
    del buff[0]
    
for word in buff:
    if word in sent_dict.keys():
        print(word, ":", sent_dict[word])
    else:
        print(word, ": NOT IN DICT")

```
